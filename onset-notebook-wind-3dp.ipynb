### The ``threshold`` option

In the above onset analysis (executed with the standard ``threshold=None``), the peak flux is found during the background period with a single encounter
of a flux of 2.69e5 1/(cmÂ² sr s MeV). This is obviously wrong, so let's use the ``threshold`` option in the initialization of ``Event()`` here.

**Important: the threshold has to be given in the original units for Wind/3DP, i.e. in eV and not MeV!**

2.69e5 1/(cmÂ² sr s **MeV**) is equal to 0.269 1/(cmÂ² sr s **eV**), so let's choose ``threshold=0.1`` (in units of eV)
because we don't see any normal value above this in our time series plot (note: 10â´ 1/(cmÂ² sr s **MeV**) = 0.1 1/(cmÂ² sr s **eV**)).

```python
event_onset_t = Event(spacecraft='wind', sensor='3dp', data_level='l2', species='e', start_date=datetime.date(2021,10,28), end_date=datetime.date(2021,10,30), data_path=path, threshold=0.1)
flux_series_t, onset_stats_t, onset_found_t, peak_flux_t, peak_time_t, fig_t, bg_mean_t = event_onset_t.analyse(viewing='2', bg_start=0, bg_length=5, channels=[3], resample_period=None, yscale='log', cusum_window=30)
```
